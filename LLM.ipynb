```python
import os
from langchain_groq import ChatGroq
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain, SimpleSequentialChain

# Step 1: Set Groq API key
os.environ["GROQ_API_KEY"] = "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"  # Replace with your actual key or use environment variable securely

# Step 2: Initialize Groq LLM
llm = ChatGroq(model_name="llama3-8b-8192", temperature=0.7)

# Step 3: Define first prompt - restaurant name
prompt_template_name = PromptTemplate(
    input_variables=['cuisine'],
    template="I want to open a restaurant for {cuisine} food. Suggest a fancy name for it."
)

name_chain = LLMChain(
    llm=llm,
    prompt=prompt_template_name,
    output_key="restaurant_name"
)

# Step 4: Define second prompt - menu items
prompt_template_items = PromptTemplate(
    input_variables=['restaurant_name'],
    template="Suggest some menu items for {restaurant_name}."
)

food_items_chain = LLMChain(
    llm=llm,
    prompt=prompt_template_items,
    output_key="menu_item"
)

# Step 5: Create a sequential chain using SimpleSequentialChain (data passes sequentially)
# But since SimpleSequentialChain only passes strings, for key-based data flow use SequentialChain below

from langchain.chains import SequentialChain

chain = SequentialChain(
    chains=[name_chain, food_items_chain],
    input_variables=["cuisine"],
    output_variables=["restaurant_name", "menu_item"],
    verbose=True
)

# Step 6: Run the chain
response = chain.invoke({"cuisine": "Indian"})

# Step 7: Output results
print("Restaurant Name:", response["restaurant_name"])
print("Menu Items:", response["menu_item"])
```

```python
from langchain_groq import ChatGroq
from langchain.agents import AgentType, initialize_agent
from langchain.utilities import WikipediaAPIWrapper
from langchain.tools import tool
import os

# Set your Groq API key
os.environ['GROQ_API_KEY'] = "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"

# Initialize the Groq LLM
llm = ChatGroq(model_name="llama3-8b-8192", temperature=0.7)

# Define a Wikipedia tool using the API wrapper
@tool
def search_wikipedia(query: str) -> str:
    """Search Wikipedia for a given query."""
    wiki = WikipediaAPIWrapper()
    return wiki.run(query)

# Define a simple math tool (replacing llm-math due to LLaMA incompatibility)
@tool
def simple_calculator(expression: str) -> str:
    """A simple calculator that evaluates basic math expressions."""
    try:
        return str(eval(expression))
    except:
        return "Invalid math expression."

# Combine tools
tools = [search_wikipedia, simple_calculator]

# Initialize the agent
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# Query the agent
response = agent.run("When was Elon Musk born? What is his age right now in 2023?")
print("\nFinal Answer:", response)
```
